```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
forwards = pd.read_csv('data/forwards.csv')
midfielders = pd.read_csv('data/midfielders.csv')
defenders = pd.read_csv('data/defenders.csv')
goalkeepers = pd.read_csv('data/goalkeepers.csv')
```

```python
forwards.head(2)
```

```python
plt.figure(1)
plt.figure(figsize=(20, 10), dpi=100)
plt.scatter(forwards['acceleration'], forwards['sprint_speed'])
plt.xlabel('acceleration')
plt.ylabel('sprint speed')
plt.show()
```

```python
plt.figure(2)
plt.figure(figsize=(20, 10), dpi=100)
plt.scatter(forwards['finishing'], forwards['free_kick_accuracy'])
plt.xlabel('finishing')
plt.ylabel('free kick accuracy')
plt.show()
```

```python
plt.figure(3)
plt.figure(figsize=(20, 10), dpi=100)
plt.scatter(midfielders['dribbling'], midfielders['long_passing'])
plt.xlabel('finishing')
plt.ylabel('free kick accuracy')
plt.show()
```
