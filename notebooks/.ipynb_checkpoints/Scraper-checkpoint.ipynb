```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import csv
```

```python
base_url = 'http://sofifa.com'
players_url = base_url + '/players?offset='

# main method
def main():
    count = 0
    player_data = []
    
    while True: 
        req = requests.get(players_url + str(count))
        soup = BeautifulSoup(req.text, 'html.parser')
        tags = soup.find_all('tr')
        
        if ((len(tags)-1) != 0):
            print('PAGE:', count/100)
            print('\n')
            itertags = iter(tags)
            next(itertags)
            
            for tag in itertags:
                link = tag.find_all('a')[0].get('href')
                playerReq = requests.get(base_url + link.strip())
                playerSoup = BeautifulSoup(playerReq.text, 'html.parser')
                
                # get general outer tags
                title = playerSoup.find_all('div', 'info')[0].find_all('h1')[0].text.split('(')
                info_div = playerSoup.find_all('div', 'info')[0].find_all('span')[0].text.strip().split(' ')
                stats_div = playerSoup.find_all('div', 'stats')[0].find_all('td')
                teams_div = playerSoup.find_all('div', 'teams')[0].find_all('ul')[0].find_all('li')
                
                stats_rows = playerSoup.find_all('div', 'columns mb-20')
                stats_row_1 = stats_rows[0].find_all('div', 'column col-3')
                stats_row_2 = stats_rows[1].find_all('div', 'column col-3')
                
                attack_card = stats_row_1[0].find_all('li')
                skill_card = stats_row_1[1].find_all('li')
                movement_card = stats_row_1[2].find_all('li')
                power_card = stats_row_1[3].find_all('li')
                
                mentality_card = stats_row_2[0].find_all('li')
                defending_card = stats_row_2[1].find_all('li')
                goalkeeping_card = stats_row_2[2].find_all('li')
                
                # get specific attributes
                player_name = title[0].strip() # get player name (string)
                player_id = title[1].split(' ')[1][:-1].strip() # get player id (string)
                
                weight = float(info_div[-1][:-2].strip()) # get player weight (float)
                height = float(info_div[-2][:-2].strip()) # get player height (float)
                yob = info_div[-3][:-1].strip() # get year of birth (string)
                dob = info_div[-4][:-1].strip() # get day of birth (string)
                mob = info_div[-5][1:].strip() # get month of birth (string)
                age = int(info_div[-6].strip()) # get player age (int)
                positions = info_div[:-7] # get player positions (string list)
                
                overall_rating = float(stats_div[0].find_all('span')[0].text.strip()) # get player overall rating (float)
                potential = float(stats_div[1].find_all('span')[0].text.strip()) # get player potential (float)
                value = parseValue(stats_div[2].find_all('span')[0].text.strip()[2:]) # get player value (float, euro)
                wage = parseValue(stats_div[3].find_all('span')[0].text.strip()[2:]) # get player wage (float, euro)
                
                preferred_foot = teams_div[0].text.strip().split('\n')[-1] # get player preferred foot (string)
                intl_reputation = int(teams_div[1].text.strip().split('\n')[-1]) # get player international reputation (int)
                weak_foot = int(teams_div[2].text.strip().split('\n')[-1]) # get player weak foot (int)
                skill_moves = int(teams_div[3].text.strip().split('\n')[-1]) # get player skill moves (int)
                work_rate = teams_div[4].find_all('span')[0].text.strip() # get player work rate (string)
                body_type = teams_div[5].find_all('span')[0].text.strip() # get player body type (string)
                
                # get attack skills
                crossing = int(attack_card[0].find_all('span')[0].text.strip())
                finishing = int(attack_card[1].find_all('span')[0].text.strip())
                heading_accuracy = int(attack_card[2].find_all('span')[0].text.strip())
                short_passing = int(attack_card[3].find_all('span')[0].text.strip())
                volleys = int(attack_card[4].find_all('span')[0].text.strip())
                
                # get general skills
                dribbling = int(skill_card[0].find_all('span')[0].text.strip())
                curve = int(skill_card[1].find_all('span')[0].text.strip())
                free_kick_accuracy = int(skill_card[2].find_all('span')[0].text.strip())
                long_passing = int(skill_card[3].find_all('span')[0].text.strip())
                ball_control = int(skill_card[4].find_all('span')[0].text.strip())
                
                acceleration = int(movement_card[0].find_all('span')[0].text.strip())
                sprint_speed = int(movement_card[1].find_all('span')[0].text.strip())
                agility = int(movement_card[2].find_all('span')[0].text.strip())
                reactions = int(movement_card[3].find_all('span')[0].text.strip())
                balance = int(movement_card[4].find_all('span')[0].text.strip())
                
                shot_power = int(power_card[0].find_all('span')[0].text.strip())
                jumping = int(power_card[1].find_all('span')[0].text.strip())
                stamina = int(power_card[2].find_all('span')[0].text.strip())
                strength = int(power_card[3].find_all('span')[0].text.strip())
                long_shots = int(power_card[4].find_all('span')[0].text.strip())
                
                aggression = int(mentality_card[0].find_all('span')[0].text.strip())
                interceptions = int(mentality_card[1].find_all('span')[0].text.strip())
                positioning = int(mentality_card[2].find_all('span')[0].text.strip())
                vision = int(mentality_card[3].find_all('span')[0].text.strip())
                penalties = int(mentality_card[4].find_all('span')[0].text.strip())
                composure = int(mentality_card[5].find_all('span')[0].text.strip())
                
                # get defense skills
                marking = int(defending_card[0].find_all('span')[0].text.strip())
                standing_tackle = int(defending_card[1].find_all('span')[0].text.strip())
                sliding_tackle = int(defending_card[2].find_all('span')[0].text.strip())
                
                # get goalkeeping skills
                gk_diving = int(goalkeeping_card[0].find_all('span')[0].text.strip())
                gk_handling = int(goalkeeping_card[1].find_all('span')[0].text.strip())
                gk_kicking = int(goalkeeping_card[2].find_all('span')[0].text.strip())
                gk_positioning = int(goalkeeping_card[3].find_all('span')[0].text.strip())
                gk_reflexes = int(goalkeeping_card[4].find_all('span')[0].text.strip())
                
                print(player_name)
                
                player_data.append([player_id, player_name, mob, dob, yob, positions, body_type, work_rate, preferred_foot, age, weight, height, overall_rating, potential, value, wage, intl_reputation, weak_foot, skill_moves, crossing, finishing, heading_accuracy, short_passing, volleys, dribbling, curve, free_kick_accuracy, long_passing, ball_control, acceleration, sprint_speed, agility, reactions, balance, shot_power, jumping, stamina, strength, long_shots, aggression, interceptions, positioning, vision, penalties, composure, marking, standing_tackle, sliding_tackle, gk_diving, gk_handling, gk_kicking, gk_positioning, gk_reflexes])
            count = count + 100
        else:
            break
    writeToCSV(player_data, 'players')
            
# method to write data to CSV file
def writeToCSV(fieldsList, title):
    csvFile = open(title + '.csv', 'w', newline='')
    try:
        writer = csv.writer(csvFile, delimiter=',')
        writer.writerow(['player_id', 'player_name', 'mob', 'dob', 'yob', 'positions', 'body_type', 'work_rate', 'preferred_foot', 'age', 'weight', 'height', 'overall_rating', 'potential', 'value', 'wage', 'intl_reputation', 'weak_foot', 'skill_moves', 'crossing', 'finishing', 'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve', 'free_kick_accuracy', 'long_passing', 'ball_control', 'acceleration', 'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power', 'jumping', 'stamina', 'strength', 'long_shots', 'aggression', 'interceptions', 'positioning', 'vision', 'penalties', 'composure', 'marking', 'standing_tackle', 'sliding_tackle', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes'])
        for row in fieldsList:
            writer.writerow(row)
    finally:
        csvFile.close()
        
# method to parse out monetary values for wage and value
def parseValue(str_val):
    if (str_val[-1] == 'M'):
        return float(str_val[:-1]) * 1000
    elif (str_val[-1] == 'K'):
        return float(str_val[:-1]) * 1000000
    else:
        return str_val
        
main()
```
